```python
import scipy
from scipy import interpolate
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt
```

```python
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
time = np.linspace(0, 11, 12)
print(speed)
print(time)
```

```python
plt.plot(speed)
plt.xlim(0, 11)
plt.ylim(0, 130)
plt.grid()
plt.show()
```

```python
f = interpolate.interp1d(time, speed, kind='cubic')
more_time = np.linspace(0, 25, 10000)
plt.plot(more_time, f(more_time))
plt.xlim(0, 11)
plt.ylim(0, 130)
plt.grid()
plt.show()
s = integrate.quad(f, 0, 11 )
print(f'S = {s[0]}')
```

```python
f = interpolate.interp1d(time, speed, kind='quadratic')
more_time = np.linspace(0, 11, 10000)
plt.plot(more_time, f(more_time))
plt.xlim(0, 11)
plt.ylim(0, 130)
plt.grid()
plt.show()
s = integrate.quad(f, 0, 11 )
print(f'S = {s[0]}')
```
